# Python Ecosystem

* Interpreter  (https://docs.python.org/3/reference/index.html)
    * Written in C

* Standard Library (Included with the language)
  (https://docs.python.org/3/library/index.html)
    * Built-in Functions, Built-in Constants, Built-in Exceptions
      * Also written in C
      * `print()`, `len()`, `range()`
      * `True`, `False`, `None`
      * `IndexError`, `KeyError`, `ValueError`

    * Python modules that you `import`

      ```python
      import datetime
      from copy import copy
      ```

* Community Developed Packages
    * Collections of Python modules that we install so that we can `import` things from them
    * Examples: `xarray`, `numpy`, `matplotlib`

# Python Modules

* Python code in a `.py` file
* Usual contents:
  * `import` statements
  * maybe constant definitions
  * maybe class definitions (`class`)
  * function definitions (`def`)
* A module is executed when it is imported
  * Execution of function and class definitions are what makes them available 
    to be called
    
Example: `salishsea_tools.viz_tools.py`

# Python Packages

* Collection of Python modules with some metadata
* Mechanism for distributing Python code among users
* Distribution via the Internet:
  * Pixi uses package channels; e.g. `conda-forge`
  * uv uses package indices; e.g. PyPI
  * code repository clones from GitHub, Bitbucket, GitLab, ...
  
### Aside

[Docs about how MOAD Python packages are structured, and why](https://ubc-moad-docs.readthedocs.io/en/latest/python_packaging/pkg_structure.html)

## Packages Give Us 2 Features/Challenges 

1. Can include Python extensions written in C
     * NumPy, SciPy, Pandas, ...
     * Compiler(s), libraries, build tools, etc. are required to install from source code
2. Packages can depend on other packages; e.g. SciPy and Pandas both require that NumPy is also installed in order to work
     * Code sharing and re-use; "standing on the shoulders of giants"
     * Leads to a web of dependencies
     * Need to construct and solve a graph to satisfy package version constraints

# Package Managers

## Why?
Because `sys.path.append(...)` doesn't scale 😱

## What Do They Do?
* Download packages we want to install,
  and their dependencies,
  from the Internet
* Store the package files where our Python code can find them
* Additional feature:
  * Isolate package files in places that don't require special permissions,
    and that don't break the operating system by over-writing other versions; avoiding problems from commands you might see on old docs:
      * `pip install` ... permission denied
      * ~~`sudo pip install`~~ 😱

# Recent History of Package Managers

## pip 👉 uv

## conda 👉 pixi

#### Others past, and present

## pip

* **p**ip **i**nstalls **p**ackages
* But only from source code
    * Until early 2020, when "wheels" were introduced
* Naive dependency resolver
    * Until late 2020: "new resolver"
* Package isolation is not part of `pip`; it is separate, but highly recommended

## conda

Scientific Python community,
lead by Travis Oliphant in 2012,
couldn't wait for the Python Packaging Authority (PyPA)'s plans for pip,
wheels,
and a sophisticated dependency resolver to come to fruition

Scientific software is written in many programming languages.
That lead to conda also being language-agnostic.

## conda

* Packages are built by maintainer, not users
    * Solves the build problem for C extensions
    * Allows installation of binary packages that aren't even Python; e.g. gfortran
    * Allows installation of different versions of Python itself
* Meta-packages; e.g. anaconda
* Dependency resolver looks at packages being installed *and* packages already installed
    * But dependency resolution is still a hard problem...
* Implicitly uses environments to isolate collections of packages
* `pip` can be used inside `conda`-managed environments

## Pixi

* An evolution of `conda` (or `conda` on steroids)
    * conda-forge or other conda channels are 1st choice source for packages
    * PyPI or other package indexes are used as a last resort
* Higher level of abstraction
    * More automatic management of package lists and environments in manifest file
* Move packages and environments information to a more visible place

## Pixi Terminology

* Workspace:
    * a directory tree that contains a manifest file, a lock file, and (eventually) a `.pixi/` directory
    * For us, usually cloned from GitHub
    * e.g. `analysis-doug/`, `SalishSeaCmd/`, `Reshapr/`
* Manifest File
* Lock File
* `.pixi/` Directory
* Mulitple Environments


## Pixi Terminology

* Workspace
* Manifest File:
    * metadata that describes the workspace
    * named `pixi.toml` (or `pyproject.toml` in packages)
    * replaces `environment.yaml` files we used with Conda
    * drives dependency resolution
    * managed by Pixi commands like `pixi add ...`
    * you need to always commit changes that Pixi makes
* Lock File
* `.pixi/` Directory
* Mulitple Environments


## Pixi Terminology

* Workspace
* Manifest File
* Lock File:
    * `pixi.lock`
    * lists the exact dependencies that were arrived at during resolution
    * provides exact reproducibility
    * don't ever edit
    * you need to always commit changes that Pixi makes
* `.pixi/` Directory
* Multiple Environments

## Pixi Terminology

* Workspace

* `.pixi/` Directory:

* Multiple Environments

## `pixi shell`

* If you doing a lot of command-line work and don't want to type `pixi run` for every command

* If you Like the conda-style `(env-name)$ ` prompt

* `pixi shell` or `pixi shell -e env-name` starts a sub-shell with the environment activated

* Use `exit` to leave the sub-shell

## Using Different Environments

* Use the `-e` or `--environment` option to run in an environment other than `default`

* `pixi run -e pkgs-envs jupyter lab pkgs-envs/PythonPkgsEnvsSlides-2026.ipynb`

* More about environments below...

## Import a Conda Environment

If `erddap-obs-matching/environment.yaml` describes an environment called `erddap-obs-matching`:

* `pixi import erddap-obs-matching/environment.yaml`

    * adds a Pixi feature called `erddap-obs-matching`
      
    * adds channels and dependencies from `environment.yaml` file to the feature
 
    * adds the feature to a Pixi environment called `erddap-obs-matching`
 
* `pixi install -e erddap-obs-matching`

* Cleanup:

    * `rm erddap-obs-matching/environment.yaml`

    * `conda env remove -n erddap-obs-matching`

* Commit and push `pixi.toml`, `pixi.lock`, and removal of `erddap-obs-matching/environment.yaml`


## Add an New Environment

* `pixi add --feature pkgs-envs python jupyter jupyterlab_rise`

* `pixi workspace environment add -f pkgs-envs --no-default-feature pks-envs`

* `pixi install -e pkgs-envs`

* Commit and push `pixi.toml` and `pixi.lock`

* `pixi run -e pkgs-envs jupyter lab`


## Clean and Recreate Environments

* `pixi clean` removes all environments
* `pixi clean -e pkgs-envs` removes the `pkgs-envs` environment

* `pixi install` creates the default environment
* `pixi install -e pkgs-envs` creates the `pkgs-envs` environment
* `pixi run -e pkgs-envs ...` also creates the `pkgs-envs` environment before running the command

## Pixi Tasks

* Command aliases

* `pixi task add -f docs --cwd docs/ docs make clean html`

* lets us do `pixi run docs` to build project docs
